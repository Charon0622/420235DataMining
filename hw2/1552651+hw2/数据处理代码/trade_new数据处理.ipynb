## 1 对trade.csv进行处理

### 1.1 读文件

```python
trade_df = pd.read_csv('../data/trade_new.csv')
```

```python
trade_df.rename(columns={'sldatime':'sldat'}, inplace = True)
```

```python
trade_df = trade_df.sort_values(by = ["vipno", "sldat", "uid", "id"])
```

```python
trade_df = trade_df.reset_index()
```

```python
trade_df.pop('index')
```

```python
trade_df
```

### 1.2 选出需要的列

```python
trade_df = trade_df[['vipno', 'uid', 'sldat', 'id', 'pluno', 'dptno', 'bndno']]
```

```python
trade_df['id'] = trade_df['id'].astype('int')
```

```python
trade_df
```

### 1.3 统计每个UID的ID数量

```python
trade_id_count = trade_df[['uid', 'id']].groupby('uid').count()
```

```python
trade_id_count = trade_id_count.reset_index()
```

```python
trade_id_count
```

### 1.4 统计每个VIPNO交易单数

```python
trade_uid_df = trade_df[['vipno', 'uid']]
```

```python
trade_uid_df
```

```python
trade_uid_df = trade_uid_df.drop_duplicates(['vipno', 'uid'])
```

```python
trade_uid_df
```

### 1.5 合并

```python
trade_count_df = pd.merge(trade_uid_df, trade_id_count, on = 'uid')
```

```python
trade_count_df
```

### 1.6 统计每个VIPNO的交易单数

```python
trade_uid_count = trade_count_df[['vipno', 'uid']].groupby('vipno').count()
```

```python
trade_uid_count
```

### 1.7 计算每个VIPNO的交易单数的60%和40%

```python
trade_uid_count['60%'] = round(trade_uid_count['uid'] * 0.6)
trade_uid_count['60%'] = trade_uid_count['60%'].astype('int')
```

```python
trade_uid_count
```

```python
trade_uid_count['40%'] = trade_uid_count['uid'] - trade_uid_count['60%']
```

```python
trade_uid_count
```

### 1.8 分割训练集和测试集

```python
trade_train_df = DataFrame(columns = trade_df.columns)
trade_test_df = DataFrame(columns = trade_df.columns)
```

```python
cur_uid = 0
cur_id = 0
for index, row in trade_uid_count.iterrows(): 
    tr_uid = row['60%']
    te_uid = row['40%']
    id_count = 0
    for i in range(cur_uid, cur_uid + tr_uid):
        id_count += trade_count_df.iat[i, 2]
    trade_train_df = trade_train_df.append(trade_df.loc[cur_id: cur_id + id_count - 1])
    cur_uid += tr_uid
    cur_id += id_count
    id_count = 0
    for i in range(cur_uid, cur_uid + te_uid):
        id_count += trade_count_df.iat[i, 2]
    trade_test_df = trade_test_df.append(trade_df.loc[cur_id: cur_id + id_count - 1])
    cur_uid += te_uid
    cur_id += id_count
```

```python
trade_train_df
```

```python
trade_test_df
```

### 1.8.1 PLUNO训练集

```python
trade_train_pluno_list = []
```

```python
cur_id = 0
for index, row in trade_count_df.iterrows(): 
    global trade_train_pluno_list
    id_count = row['id']
    cur_set = set()
    for i in range(cur_id, cur_id + id_count):
        if not math.isnan(trade_train_df.iat[i, 4]):
            cur_set.add(trade_train_df.iat[i, 4])
    cur_id += id_count
    if len(cur_set) != 0:
        trade_train_pluno_list.append(list(cur_set))
```
